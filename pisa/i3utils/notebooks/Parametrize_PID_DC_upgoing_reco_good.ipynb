Define all functions that will be used for fits

```python
# Fermi-Dirac-like step
FD_step = lambda E, level, step, width, offset: level/(np.exp((step - np.log10(E))/width) + 1.) + offset
FD_p0 = [0.5, 0.8, 0.2, 0.01]
FD_step_str = '%.2e/(np.exp((%.2e - np.log10(E))/%.2e) + 1.) + %.2e'

# generic exponential
exp_func = lambda E, scale, timeconst, offset: scale*np.exp(-E/timeconst) + offset
exp_p0 = [0.5, 10., 0.1]
exp_str = '%.2e*np.exp(-E/%.2e) + %.2e'

# gaussian (for the slope)
gauss_func = lambda E, scale, pos, width, offset: scale*norm(loc=pos, scale=width).pdf(E) + offset
gauss_p0 = [1., 10., 5., 0.02]
gauss_str = '%.2e*scipy.stats.norm(loc=%.2e, scale=%.2e).pdf(E) + %.2e'

# gaussian in log(E)
gauss_log_func = lambda E, scale, pos, width, offset: scale*norm(loc=pos, scale=width).pdf(np.log10(E)) + offset
gauss_log_p0 = [1., 10., 5., 0.02]
gauss_log_str = '%.2e*scipy.stats.norm(loc=%.2e, scale=%.2e).pdf(np.log10(E)) + %.2e'
```

```python
func = gauss_log_func
func_str = gauss_log_str
first_guess = gauss_log_p0

save_dir = '/Users/feifeihuang/pisa_analysis/notebooks/PID_DC12_reco/'

param_dict = {}
data_dict = {
             'nue':  {  'label': r'$\nu_e$',
                        'ID':    NuEfracID},
             'numu': {  'label': r'$\nu_\mu$',
                        'ID':    NuMufracID},
             'nutau':{  'label': r'$\nu_\tau$',
                        'ID':    NuTaufracID},
             'NC':   {  'label': r'$NC$',
                        'ID':    NCfracID}
            }

#figure(figsize=(8,5))
plt.figure(figsize=(12,8))
colors = ['b','g','r','c']
flavors = [r'$\nu_\mu$','NC',r'$\nu_e$',r'$\nu_\tau$']
for i, (key, val) in enumerate(data_dict.items()):
    
    plt.subplot(2, 2, 1+i)
    
    #title(val['label'], size='xx-large')
    plt.plot(energybins, val['ID'],linestyle='',marker='.')
    
    plt.grid()
    plt.xlim(0,80)
    plt.ylim(ymin=0)
    if i>1:
        plt.xlabel('Energy [GeV]', size='x-large')
    if not i%2:
        plt.ylabel('Track ID probability', size='x-large')
    
    try:
        
        ############### TCA Addition ##################
        if(key in ['numu']):
            func = FD_step
            func_str = FD_step_str
            first_guess = FD_p0
        elif (key in ['nutau']):
            func = exp_func
            func_str = exp_str
            first_guess = exp_p0
            #func = gauss_log_func
            #func_str = gauss_log_str
            #first_guess = [-1., 10., 3., 0.1]
            #func = gauss_func
            #func_str = gauss_str
            #first_guess = [-1.4, 10., 3., 0.1]
        elif (key in ['nue','NC']):
            func = gauss_func
            func_str = gauss_str
            first_guess = [-1., 10., 3., 0.1]
            #func = exp_func
            #func_str = exp_str
            #first_guess = exp_p0
            #func = gauss_log_func
            #func_str = gauss_log_str
            #first_guess = gauss_log_p0
        elif (key == ''):
            func = gauss_func
            func_str = gauss_str
            first_guess = gauss_p0
        else:
            func = gauss_log_func
            func_str = gauss_log_str
            first_guess = gauss_log_p0
        #####################################################
                
        fit, cov = curve_fit(func, energybins, val['ID'], p0=first_guess)
        plt.plot(energybins, func(energybins, *fit), '-'+colors[i],label=flavors[i])

        plt.legend(loc='best')
        #grid()
        
        stringrepr = func_str%tuple(fit)
        param_dict[key] = {'trck': 'lambda E: '+stringrepr,
                           'cscd': 'lambda E: 1. - (%s)'%stringrepr}
    except RuntimeError:
        print "something wrong"
        pass

#legend(loc='best')
#grid()
#xlim(0,50)
print param_dict
plt.savefig(figDir+'PID_up_1X72.png')
```
